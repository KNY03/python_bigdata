```python
# 리스트 관련 함수
## 리스트 요소 추가(append())

a = [1, 2, 3]
a.append(4)
a
```

```python
# 리스트 정렬 - sort()
a = [1, 4, 3, 2]
a.sort()
a

a = ["a", "c", "b", "d"]
a.sort(reverse=True)
a
```
