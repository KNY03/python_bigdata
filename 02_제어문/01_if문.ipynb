```python
money = True

if money: 
    print("택시를 타고 간다.")
else: 
    print("걸어간다.")
```

```python
# 학점이 90점 이상이면 A
# 80점 이상 90점 미만이면 B
# 70점 이상 80점 미만이면 C
# 60점 이상 70점 미만이면 D
# 60점 미만이면 F

score = int(input("점수를 입력하세요: "))

if score >= 90:
    grade = "A학점"
    print(f"당신의 입력값은 {score}")
elif (score >= 80 and score < 90):
    grade = "B학점"
    print(f"당신의 입력값은 {score}") 
elif (score >= 70 and score < 80):
    grade = "C학점"
    print(f"당신의 입력값은 {score}") 
elif (score >= 60 and score < 70):
    grade = "D학점"
    print(f"당신의 입력값은 {score}") 
else:
    grade = "F학점"
    print(f"당신의 입력값은 {score}") 

print(f"당신의 학점은 {grade}입니다.")
```
