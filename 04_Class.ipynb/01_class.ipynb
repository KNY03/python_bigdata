```python
result1 = 0

def add1(num):
    global result1
    result1 += num

    return result1

print(add1(3))
print(add1(4))
print(add1(5))
```

```python
result2 = 0

def add2(num):
    global result2
    result2 += num

    return result2

print(add2(5))
print(add2(10))
print(add1(20))
```

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result = self.result + num
        return self.result


cal1 = Calculator()
cal2 = Calculator()
print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(5))
print(cal2.add(10))
```

```python
# 더하기 기능

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
a = FourCal()
a.setdata(4, 3)
print(a.first)
print(a.second)
print(a.add())

b = FourCal()
b.setdata(3, 7)
```

```python
# 곱하기, 빼기, 나누기 기능 만들기
## mul() - 곱하기
## sub() - 빼기
## div() - 나누기

### 계산기 2개 만들기 (cal1, cal2)
### cal1(10, 20) 
### cal2(20, 5)

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
        return self

    def add(self):
        result = self.first + self.second
        return result
     
    def mul(self):
        result = self.first * self.second
        return result
     
    def sub(self):
        result = self.first - self.second
        return result
     
    def div(self):
        result = self.first / self.second
        return result

cal1 = FourCal()
cal1.setdata(10, 20)
print(cal1.add())
print(cal1.mul())
print(cal1.sub())
print(cal1.div())

print("")

cal2 = FourCal()
cal2.setdata(20, 5)
print(cal2.add())
print(cal2.mul())
print(cal2.sub())
print(cal2.div())

print("")

print(FourCal().setdata(10, 20).add())

```

```python
class MoreFurCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result 
    
cal4 = MoreFurCal()
cal4.setdata(2, 4)
cal4.pow()
```

```python
# 매서드 오버라이딩

class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
        
cal1 = FourCal()
cal1.setdata(4, 0)
# cal1.div()

cal2 = SafeFourCal()
cal2.setdata(4, 0)
cal2.div()
```
