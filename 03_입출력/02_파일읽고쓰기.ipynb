```python
f = open("./data/새파일.txt", "w")

for i in range(1, 11):
    data = f"{i}번째 줄 입니다.\n"
    f.write(data)

f.close()
```

```python
f = open("data/새파일.txt", "r")
while True:
    line = f.readline()
    if not line: break
    print(line)

f.close()
```

```python
# readlines() 함수 이용하기
f = open("data/새파일.txt", "r")
lines = f.readlines()
print(lines)
f.close()
```
