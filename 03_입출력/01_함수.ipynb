```python
# 함수란 무엇인가?
## 믹서기, 자판기

# 파이썬 함수의 구조

# def 함수이름(매개변수):
#     수행할 문장1
#     수행할 문장2
#     ...


def add(a, b):
    return a + b

result = (add(3, 4))
print(result)
```

```python
# 일반적인 함수
## 입력값: O, 반환값: O

# def 함수이름(매개변수):
#     수행할 문장1
#     수행할 문장2
#     ...
#     return 반환값

# 예제
def add(a, b):
    result = a + b
    return result

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수
## 입력값: X, 반환값: O

def say():
    return 'Hi'

a = say()
print(a)
```

```python
# 반환값이 없는 함수
## 입력값: O, 반환값: X

def add(a, b):
    print(f"{a} + {b} = {a + b}")

a = add(3, 4)
print(a)
```

```python
# 입력값도 없고 반환값도 없는 함수
## 입력값: X, 반환값: X

def say():
    print('Hi')

a = say()
print(a)
```

```python
# 매개변수 지정하여 호출하기
def sub(a, b):
    return a - b

result1 = sub(7, 3)
print(result1)

result2 = sub(b=7, a=3)
print(result2)
```

```python
# # 입력값이 몇개가 될지 모를 때는 어떻게 할까?

# def 함수이름(*args):
#     수행할 문장1
#     수행할 문장2
#     ...

def add_many(*args): # args = [1, 2, 3, 4, 5]
    result = 0
    for i in args:
        result += i
    return result

a = add_many(1, 2, 3, 4, 5)
print(a)
```

```python
#  키워드 매개변수, kwargs
## 키워드값의 형태로 매개변수를 받을 때
## 매개변수 앞에 **(2개)를 붙임

def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(a=1)
print_kwargs(name='foo', age=3)
```

```python
def create_profile(**profiles):
    print("===== Profile =====")
    for key, value in profiles.items():
        print(f"{key}: {value}")

create_profile(name='Alice', age=30, city='New York', job='Engineer', hobby='Photography')
```

```python
# *args와 **kwargs 혼용
def mixed_function(name, *args, **kwargs):
    print(f"Name: {name}")
    print("Args(추가인수):", args)
    print("Kwargs(키워드 추가 인수):", kwargs)

mixed_function('Bob', 1, 2, 3, 4, 5, age=30, city='서울', job='Designer')
```

```python
def add_and_multiply(a, b):
    return a + b, a * b

result = (add_and_multiply(3, 4))
print(result)
```

```python
def say_nick(nick):
    if nick == '바보':
        return
    print(f"나의 별명은 {nick}입니다.")

# say_nick("야호")
say_nick("바보")
```

```python
# 매개변수 초기값 미리 설정하기
def say_myself(name, age, man=True):
    print(f"나의 이름은 {name}입니다.")
    print(f"나이는 {age}살입니다.")
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")

result_m = say_myself("홍길동", 30)
print("")
result_w = say_myself("김영희", 25, False)
```

```python
# lambda 여약어
# def 함수이름(매개변수):
#    수행할 문장1
#    return

# 함수이름 = lambda 매개변수1, 매개변수2, ...: 수행할 문장

# def add(a, b):
#     return a + b

# lambda 
# 함수이름 = lambda 매개변수1, 매개변수2: 수행할 문장
add = lambda a, b: a + b
result = add(3, 4)
print(result)
```

```python
def add(a, b):
    """
    함수의 독스트링(Docstring)
    두 숫자를 더하는 함수
    설명: 다른사람 또는 나를 위한 문서
    """
    return a + b

result = add(3, 4)
print(result)
print(add.__doc__)
```
