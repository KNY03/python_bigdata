```python
#abs()
abs(-3) #절대값
```

```python
# all() - 인수의 요소가 모두 참ㅁ이면 참으로 반환
all([1, 2, 3])
```

```python
# any() 참이 하나라도 있으면 참
any([1, 2, 3, ""])
```

```python
# dir -> 객체가 지닌 변수나 함수를 보여주는 함수
dir([1, 2, 3])
```

```python
# id -> 객체를 입력받아 객체의 고유 주소값을 반환하는 함수
a = 3
b = 4
print(id(a))
print(id(b))
```

```python
# map(f, iterable)은 각 요소를 f()함수에 적용한 결과를 보여준다

def two_times(x):
    return x * 2


list(map(two_times, [1, 2, 3, 4]))
```
